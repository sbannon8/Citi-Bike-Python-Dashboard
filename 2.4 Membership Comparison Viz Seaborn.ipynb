## Import Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
```

## Import Data

Columns needed:
- ride_id
- start station name
- avgTemp
- date
- rideable_type
- member_casual
- started_at
- ended_at

```python
# list cols to use
use_cols = ['ride_id', 'start_station_name', 'date', 'avgTemp', 'rideable_type', 'member_casual', 'started_at', 'ended_at']

# import data
df = pd.read_csv('ny_data.csv',
usecols=use_cols, # import only specified columns
parse_dates=['started_at', 'ended_at', 'date'], # set these col as datetime dtype
low_memory=False)
```

```python
df = df[(df['date'] >= '2022-01-01') & (df['date'] <= '2022-12-31')]
```

## Visualizations

#### Column Chart with Matplotlib

```python
df['rideable_type'].value_counts().plot(kind = 'bar')
```

#### Column Chart with Seaborn

```python
sns.countplot(data = df, x = 'rideable_type')
plt.title("Ride types count")
plt.show()
```

### Top 20 Starting Stations ----

#### (Bar Chart)

```python
# group by start station name and count the occururences for each
df_t = df.groupby('start_station_name').size().reset_index(name='count')
# create list with only the top 20 from above
top20 = df_t.nlargest(20, 'count')
top20
```

```python
# create bar chart of top 20
sns.barplot(data = top20, x = 'count', y = 'start_station_name', hue ='start_station_name', palette=("Blues_r"), legend=False)
plt.title("Most frequent start stations")
plt.show() 
```

### Ride Count vs Temp ----

#### (Dual Axis Line Plot)

```python
# first create new df that groups data by day for the daily ride count & avg temp
daily_df = df.groupby('date').agg({
    'ride_id':'count', # the daily ride count
    'avgTemp':'mean' #the average daily temperature
}).reset_index()
```

```python
daily_df.head(3)
```

```python
fig,ax = plt.subplots(figsize=(10,5)) # set size

# 1st x-axis: DAILY RIDES
sns.lineplot(data=daily_df['ride_id'], color="b") # set data & color
ax.set_xlabel("Year 2022", fontsize=14) # set x-axis label
ax.set_ylabel("Bike rides daily", color = "navy", fontsize = 14) # set y-axis label

# 2nd x-axis: AVG TEMP
ax2 = ax.twinx() # create dual axis using x-axis
sns.lineplot(data = daily_df['avgTemp'], color="r", ax=ax2) # set data & color
ax2.set_ylabel("Average temperatures", color="red", fontsize=14) # set x_axis label

plt.title("Temperature and trips in 2022", color="red", fontsize=14) # set title
plt.show()
```

### Trip Duration by Member Status ---

#### (Box Plot)

```python
# create new column that stores the trip duration
df['trip_duration_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
df.head(3)
```

```python
# create box plot showing the trip duration by member status
colors= {'member':'blue', 'casual':'cyan'}
sns.boxplot(
    data=df,
    x='member_casual',
    y='trip_duration_min',
    hue='member_casual',
    palette=colors,
    legend=False
)
```

```python
# show number of trips over 1 day
over24hrs_count = df[df['trip_duration_min'] > 1440].shape[0]
print(over24hrs_count)
```

```python
# store trips over 1 day into new df
over24hrs_trips = df[df['trip_duration_min'] > 1440]
over24hrs_trips.head()
```

```python
df['ride_id'].nunique()
```

```python
39772 / 29838166
```

**Insight:** 0.13% of rides have trip durations of more than 24 hours. According to the Citi Bikes website, all rides must be returned after 30 min to avoid extra fees (even with the 24 hour pass, it is 24 hours of unlimited 30 min, with extra fees applying if a ride exceeds the 30 min). Therefore, these data points could point to unreturned, stolen, or lost bikes or technical issues.

Accoring the the box plot, casual customers have a much higher rate of rides over 24 hours, with rides lasting as long as over 5 months! This is an area that could be explored more - possibly pointing to a higher risk of non-returned bikes among casual customers vs members.

**Show for only trips <= 1 hour**

```python
# store trips UNDER 1 day into new df
trips_1hr = df[df['trip_duration_min'] <= 60]
```

```python
# create box plot showing the trip duration by member status for trips UNDER 24 hrs
colors= {'member':'blue', 'casual':'cyan'}
sns.boxplot(
    data=trips_1hr,
    x='member_casual',
    y='trip_duration_min',
    hue='member_casual',
    palette=colors,
    legend=False
)
```

**Insight:** Members typically take shorter trips then casual riders. Casual riders have a median of about 2 - 5 minutes higher then members as well as a slighter larger data range.

```python
# show number of trips <= 0 min
under0min_count = df[df['trip_duration_min'] <= 0].shape[0]
print(under0min_count)
```

```python
# store trips <= 0 min into new df
trips_0min = df[df['trip_duration_min'] <= 0]
trips_0min.head()
```

**Insight:** There are 357 instances of trips that are 0 minutes or less. This could be due to technical errors - these could be investigated further. If trips are tracked via modules on the bike, the bike could be flagged to see if it is continually outputting errors and needs to be repaired.

#### (Violin Plot)

```python
# 'remove' the < 0 min entries from our 1 hour df
trips_0to60min = trips_1hr[trips_1hr['trip_duration_min'] > 0]
```

```python
sns.violinplot(
    data=trips_0to60min,
    x='member_casual',
    y='trip_duration_min',
    hue='member_casual',
    palette=colors,
    legend=False
)
```

```python
# output the median trip duration of casual riders
trips_0to60min[trips_0to60min['member_casual'] == 'casual']['trip_duration_min'].median()

# this number will be higher than graphed as it includes all data points (not just those within the 1 hour range)
```

**Insights:** The member category has a pretty quick taper after it's median; whereas the casual category has a more gradual taper. This again highlights that member's have a more predictable trip duration with most taking quick 5 min rides where casual riders don't all behave quite as uniformly, with slightly longer rides on average. There is a slight density bump in casual ries at 30 min which aligns with Citi Bikes 30 min maximum; however it is not a signifcant density increase, signifying that most keep their trip under the maximum allowance (typically keeping trips to about 8 minutes).

**NOTE** These medians are reflective of only trips within an hour - this median will be higher with the inclusion of all points. For instance, the true median of casual riders is about 14 min (pushed about 6 min higher than the <= 1 hour trips median).

#### (Strip Plot)

```python
sns.stripplot(
    data=trips_0to60min,
    x='member_casual',
    y='trip_duration_min',
    hue='rideable_type', # color code based on ride type (classic vs electric)
    dodge=True
)
plt.legend(loc='upper right') 
plt.show()
```
